```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas import get_dummies
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
from sklearn.preprocessing import MinMaxScaler
```

```python
training_set = pd.read_csv('train.csv')
training_set.head()
```

```python
test_set = pd.read_csv('test.csv')
test_set.head()
```

```python
training_set.info()
```

```python
training_set.columns[-1]
```

```python
test_set[training_set.columns[-1]] = 0
```

```python
test_set
```

```python
dataset = pd.concat([training_set, test_set])
```

```python
dataset.info()
```

```python
dataset.isnull().sum().sort_values(ascending=False)
```

```python
dataset.drop(['PoolQC','YrSold','SaleType','SaleCondition','YrSold','MoSold'],axis =1, inplace=True)
```

```python
#df['MiscFeature'].value_counts()
def special_return_bin (Feature):
    if pd.isnull(Feature):
        return 0
    else:
        return 1
    
    
dataset['MiscFeature'] = dataset['MiscFeature'].apply(lambda x: special_return_bin(x))
```

```python
dataset['Alley'] = dataset['Alley'].apply(lambda x: special_return_bin(x))
```

```python
def special_return_value (Feature, Name):
    if pd.isnull(Feature):
        return ('No_'+ Name)
    else:
        return Feature
    
dataset['Fence'] = dataset['Fence'].apply(lambda X: special_return_value(X, 'Fence'))

dummies = pd.get_dummies(dataset['Fence'], dtype=float, drop_first=True)
dataset = pd.concat([dataset.drop('Fence', axis=1), dummies], axis=1)
```

```python
dataset['MasVnrType'] = dataset['MasVnrType'].apply(lambda X: special_return_value(X,'Mas'))

dummies = pd.get_dummies(dataset['MasVnrType'], dtype=float, drop_first=True)
dataset = pd.concat([dataset.drop('MasVnrType', axis=1), dummies], axis=1)
```

```python
dataset['FireplaceQu'] = dataset['FireplaceQu'].apply(lambda X: special_return_value(X,'Fire'))

dummies = pd.get_dummies(dataset['FireplaceQu'], dtype=float, drop_first=True)
dataset = pd.concat([dataset.drop('FireplaceQu', axis=1), dummies], axis=1)
```

```python
def impute_lotFrontage (LotFrontage, stFlrSF):
    if pd.isnull(LotFrontage):
        return (stFlrSF*0.075934)
    else:
        return LotFrontage
    
dataset['LotFrontage'] = dataset.apply(lambda x: impute_lotFrontage(x['LotFrontage'], x['1stFlrSF']), axis = 1)
```

```python
dataset['GarageCond'] = dataset['GarageCond'].apply(lambda X: special_return_value(X, 'Garage'))

dummies = pd.get_dummies(dataset['GarageCond'], dtype=float, drop_first=True)
dataset = pd.concat([dataset.drop('GarageCond', axis=1), dummies], axis=1)
```

```python
dummies = pd.get_dummies(dataset['GarageType'], dtype=float, drop_first=True)
dataset = pd.concat([dataset.drop('GarageType', axis=1), dummies], axis=1)
```

```python
missing = dataset.isnull().sum().sort_values(ascending=False).head(24) # the list of all the missing values
missing_percentage = (missing/len(dataset))*100
missing_percentage #How are we missing almost, all the values in the datasetLotFrontage   
```

```python
missing.keys()
```

```python
dataset.drop(missing.keys(), axis = 1, inplace= True)
```

```python
len(dataset.dtypes[(dataset.dtypes != 'int64') & (dataset.dtypes != 'float64')])
```

```python
non_int = dataset.dtypes[(dataset.dtypes != 'int64') & (dataset.dtypes != 'float64')].keys().array
non_int = list(non_int)
dataset[non_int].head()
dummies = pd.get_dummies(dataset[non_int], dtype=float, drop_first=True)
dataset = pd.concat([dataset.drop(non_int, axis=1), dummies], axis=1)
```

```python
dataset.dtypes[(dataset.dtypes != 'int64') & (dataset.dtypes != 'float64')]
```

```python
train = dataset[dataset['SalePrice']!= 0]
test = dataset[dataset['SalePrice'] == 0]
```

```python
train.head()
```

```python
test.head()
```

```python
X = train.drop('SalePrice', axis = 1).values
y = train['SalePrice'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=32)
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
early_stop = EarlyStopping(monitor='val_mse', mode='min', patience=5)
```

```python
model = Sequential()

model.add(Dense(226, activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(113, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(56, activation='relu'))
model.add(Dropout(0.1))

model.add(Dense(28, activation='relu'))

model.add(Dense(14, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse', metrics=['mse'])
```

```python
model.fit(x=X_train, y=y_train,batch_size=128, epochs= 600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
test = test.drop('SalePrice', axis = 1).values
test = scaler.transform(test)
```

```python
predictions = model.predict(test)
```
